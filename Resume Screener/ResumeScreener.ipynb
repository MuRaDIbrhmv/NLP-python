Cleaning the data

```python
import re 
```

```python
def cleanResume(x):
    cleaned = re.sub(r'http\S+','', x)
    cleaned = re.sub(r'https\S+','', cleaned)
    cleaned = re.sub(r'@\S+','', cleaned)
    cleaned = re.sub(r'#\S+','', cleaned)
    cleaned = re.sub(r'â\S+','', cleaned)
   

    return cleaned
```
